Q36. Compare Loan_amount across age groups.

```python
query= """SELECT CASE
        WHEN U.Age BETWEEN 18 AND 25 THEN 'Gen Z (18-25)'
        WHEN U.Age BETWEEN 26 AND 40 THEN 'Millennials (26-40)'
        WHEN U.Age BETWEEN 41 AND 60 THEN 'Boomers (41-60)'
        ELSE 'Senior Citizens (60+)'
        END AS Age_Group,
        AVG(L.Loan_amount) AS Average_Loan_Amount FROM bpnl_users U
        JOIN bpnl_loans L ON U.User_id = L.User_id GROUP BY Age_Group"""
result()
```

Q37. Display complete user financial behavior.

Q38. Find city-wise average Loan amount and Default risk.

Q39. Display users whose Credit utilisation & Missed Payments are High.

```python
query= """SELECT U.User_id, U.Age,L.Credit_utilization_ratio, P.Missed_payments,
          P.Transactions_per_month FROM bpnl_users U JOIN bpnl_loans L
          ON U.User_id = L.User_id JOIN bpnl_payments P ON U.User_id = P.User_id
          WHERE L.Credit_utilization_ratio = 1.0 AND P.Missed_Payments = 4
          ORDER BY P.Transactions_per_month DESC"""
result()
```

Q40. Find users whose Loan amount exceeds Monthly_income.

```python
query= """SELECT U.User_id, U.Monthly_income, L.Loan_amount,
          (L.Loan_amount - U.Monthly_income) AS Difference
          FROM bpnl_users U JOIN bpnl_loans L ON U.User_id = L.User_id
          WHERE L.Loan_amount > U.Monthly_income ORDER BY Difference DESC"""
result()
```

Q41. Calculate average Credit utilization ratio by age group.

```python
query= """SELECT CASE
        WHEN U.Age BETWEEN 18 AND 25 THEN 'Gen Z (18-25)'
        WHEN U.Age BETWEEN 26 AND 40 THEN 'Millennials (26-40)'
        WHEN U.Age BETWEEN 41 AND 60 THEN 'Boomers (41-60)'
        ELSE 'Senior Citizens (60+)'
        END AS Age_Group,
        AVG(L.Credit_utilization_ratio) AS Avg_Credit_Utilization
        FROM bpnl_users U JOIN bpnl_loans L ON U.User_id = L.User_id
        GROUP BY Age_Group ORDER BY Avg_Credit_Utilization DESC"""
result()
```

Q42. Rank Top 10 users based on Loan amount.

```python
query= """SELECT U.User_id, L.Loan_amount,
          RANK() OVER (ORDER BY L.Loan_amount DESC) AS Loan_Rank
          FROM bpnl_users U JOIN bpnl_loans L
          ON U.User_id = L.User_id Limit 10"""
result()
```

Q43. Find top 10 cities with highest BNPL transaction activity.

```python
query= """SELECT U.City, SUM(P.Transactions_per_month) AS Total_Transactions
          FROM bpnl_users U JOIN bpnl_payments P ON U.User_id = P.User_id
          GROUP BY U.City ORDER BY Total_Transactions DESC LIMIT 10"""
result()
```

Q44. Calculate cumulative Loan amount Year-wise.

```python
query= """SELECT Year, SUM(Loan_amount) AS Total_Loan_Amount,
          SUM(SUM(Loan_amount)) OVER (ORDER BY Year) AS Cumulative_Loan_Amount
          FROM bpnl_loans GROUP BY Year ORDER BY Year"""
result()
```

Q45. Find percentage contribution of each Lending company to total Loan amount.

```python
query= """SELECT A.Lending_company, SUM(L.Loan_amount) AS Total_Loan,
          ROUND((SUM(L.Loan_amount) * 100.0) /
          (SELECT SUM(Loan_amount) FROM bpnl_loans), 3)
          AS Percentage_Contribution FROM bpnl_accounts A JOIN bpnl_loans L
          ON A.User_id = L.User_id GROUP BY A.Lending_company
          ORDER BY Percentage_Contribution DESC"""
result()
```

Q47. Analyze relationship between Missed payments and Default risk.

Q48. Categorize users into Low, Medium, and High Risk groups.

```python
query= """SELECT User_id, Default_risk, CASE
          WHEN Default_risk BETWEEN 1 AND 3 THEN 'Low Risk'
          WHEN Default_risk BETWEEN 4 AND 6 THEN 'Medium Risk'
          ELSE 'High Risk' END AS Risk_Category FROM bpnl_users"""
result()
```

Q49. Categorize users based on Transactions per month.

```python
query= """SELECT COUNT(*) AS Total_Users,
          ROUND(AVG(Transactions_per_month)) as Avg_Transactions_per_month, CASE
          WHEN Transactions_per_month < 10 THEN 'Low Activity'
          WHEN Transactions_per_month BETWEEN 10 AND 25 THEN 'Moderate Activity'
          ELSE 'High Activity' END AS Activity_Level FROM bpnl_payments
          GROUP BY Activity_Level"""
result()
```

Q50. Detect abnormal Credit_utilization_ratio values and identify risky users.

```python
query="""SELECT U.User_id, L.Credit_utilization_ratio, U.Default_risk
         FROM bpnl_users U JOIN bpnl_loans L ON U.User_id = L.User_id
         WHERE L.Credit_utilization_ratio = 1.0 AND U.Default_risk = 9"""
result()
```

Q51. Find users with highest transaction frequency and compare repayment behavior.

```python
query="""SELECT P.Repayment_method, COUNT(U.User_id) AS Total_Users,
         AVG(P.Transactions_per_month) AS Avg_Transactions,
         AVG(P.Missed_payments) AS Avg_Missed_Payments FROM bpnl_users U
         JOIN bpnl_payments P ON U.User_id = P.User_id
         GROUP BY P.Repayment_method ORDER BY Avg_Missed_Payments DESC"""
result()
```
